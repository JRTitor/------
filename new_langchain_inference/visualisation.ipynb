```python
for i in range(len(data_frames)):
    data_frames[i]['len'] = data_frames[i].response.apply(len)
```

```python
for i in range(len(data_frames)):
    data_frames[i]['tokens'] = data_frames[i].response.apply(lambda x: len(x.split()))
```

```python
for i in range(len(data_frames)):
    data_frames[i]['tokens_norm'] = data_frames[i].response.apply(lambda x: min(len(x.split()), 2048))
```

```python
combined_stats = dict()
```

```python
for i in range(len(data_frames)):
    stats = {
        'mean': data_frames[i][' score'].mean(),
        'var': data_frames[i][' score'].var(),
        'zeros': (data_frames[i][' score'] == 0).sum(),
        'fives': (data_frames[i][' score'] == 5).sum()
    }
    # Store the statistics for the current column
    combined_stats[FILES[i]] = stats
```

```python
for i in range(len(data_frames)):
    data_frames[i].to_csv(os.path.join('../marked', FILE_NAME[i]))
```
